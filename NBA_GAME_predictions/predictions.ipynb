## NBA GAME PREDICTIONS

### Data Cleaning

```python
import pandas as pd

df = pd.read_csv('nba_games.csv', index_col=0)
```

```python
df.head()
```

```python
df = df.sort_values("date")
```

```python
df = df.reset_index(drop=True)  # without the drop=True the old index gets added as a column to the df
df.head()
```

```python
del df["mp.1"]
del df["mp_opp.1"]
del df["index_opp"]
```

```python
df
```

```python
# The target is the next value in the win column, the result of the NEXT game
def add_target(team):
    team["target"] = team["won"].shift(-1)
    return team

# Apply the add_target function for a certain team by grouping the df by that team
df = df.groupby("team", group_keys=False).apply(add_target)
```

```python
df[df["team"] == "HOU"]
```

```python
# Replace all of the null occurrences of null values in the target column 
df['target'][pd.isnull(df['target'])] = 2
```

```python
# Turn the target into a boolean integer value
df['target'] = df['target'].astype(int, errors='ignore')
```

```python
# Check if there is a class imbalance
df["won"].value_counts()
```

```python
df["target"].value_counts()
```

```python
null_counts = df.isnull().sum()  # Check to see if any columns in the datagrame have a null value. 
null_counts = null_counts[null_counts > 0]
null_counts
```

```python
# Since these values are null almost always, we can just get rid of all of them
valid_columns = df.columns[~df.columns.isin(null_counts.index)]
```

```python
df = df[valid_columns].copy()   # Copy allows us to create a new dataframe instead of getting a part of another one.
```

### Machine Learning

Our current dataframe has a large number of columns. This is a problem because it can create a lot of noise or overfitting and mess with our results. Before going into the machine learning, we will use a feature selector that will allow us to get the important columns. 

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import RidgeClassifier

# We have time-series data, so we need to make sure that we use a split that accounts for this
rr = RidgeClassifier(alpha=1)
split = TimeSeriesSplit(n_splits=3)

sfs = SequentialFeatureSelector(rr,     # The model being used
                                n_features_to_select=30,    # The amount of columns we want left in our data
                                direction="forward",    # This will pick the best feature first and then continue to pick until we have 30
                                cv=split)   #   Time series split used for cross-validation
```

```python
# Ridge Regression works best when numbers are normalized within a range, in this case we will normalize the scores between 0 and 1
removed_columns = ["season", "date", "won", "target", "team", "team_opp"]   # These features cant really be turned into numbers, so we will remove them for now
```

```python
selected_colums = df.columns[~df.columns.isin(removed_columns)]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[selected_colums] = scaler.fit_transform(df[selected_colums])
```

```python
df
```

```python
sfs.fit(df[selected_colums], df['target'])
```

```python
predictors = list(selected_colums[sfs.get_support()])
```

```python
predictors
```

These features have been selected as the best to pick by the Feature Selector for Ridge Regression.

The one's with 'max' indicate the player on the team that shot the best.

```python
def backtest(data, model, predictors, start=2, step=1):
    # Split data by season. Uses past season to predict future seasons. 
    all_predictions = []

    seasons = sorted(data["season"].unique())

    # This loop goes from the third season onwards to the last season
    # The training data is split as all the seasons before, and the test is that season
    for i in range(start, len(seasons), step):
        season = seasons[i]

        train = data[data["season"] < season]
        test = data[data["season"] == season]

        model.fit(train[predictors], train['target'])

        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)

        combined = pd.concat([test['target'], preds], axis=1)
        combined.columns = ["actual", "prediction"]

        all_predictions.append(combined)
    
    return pd.concat(all_predictions)
```

```python
predictions = backtest(df, rr, predictors)
```

```python
predictions
```

```python
# Now let's make a metric to measure the similatirty of these numbers
from sklearn.metrics import accuracy_score

accuracy_score(predictions["actual"], predictions["prediction"])
```

```python
# Could get a little better score by removing the 2s 
from sklearn.metrics import accuracy_score

sample_predictions = predictions[predictions["actual"] != 2]
accuracy_score(sample_predictions["actual"], sample_predictions["prediction"])
```

## Improving the Model

In basketball there are many other nuances that are not accounted for fully in our previous model. The model would have much more accurate predictions if it was given better guidelines to predict on. 

For example, we need to have an accuracy that's higher than the % of home teams that win the games or else, we could just use that as our metric instead of doing ML.

```python
df.groupby("home", group_keys=False).apply(lambda x: x[x["won"] == 1].shape[0] / x.shape[0])
```

```python
df_rolling = df[list(selected_colums) + ["won", "team", "season"]]
```

```python
# We can create a weighted average of how a team performed in their last 10 games, as opposed to just the one game that we are using now
def find_team_averages(team):
    rolling = team.rolling(10).mean()
    return rolling

df_rolling = df_rolling.groupby(["team", "season"], group_keys=False).apply(find_team_averages)
```
